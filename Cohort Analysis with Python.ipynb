```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('Online_Retail.csv')
```

```python
# Aim: To present a cohort chart,each row on the chart represents 
#a user's sign up month and each numeric column represents the percent of users that are still "active".
```

```python
df = df.dropna(subset=['CustomerID'])
```

```python
df.dtypes
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
#getting the Invoice Month
import datetime as dt
def get_month(x):
    return dt.datetime(x.year, x.month, 1)
df['InvoiceMonth'] = df['InvoiceDate'].apply(get_month)
```

```python
#cohort month : the first month for the specific user ID when that user shopped at this online retailer.
df['CohortMonth'] = df.groupby('CustomerID')['InvoiceMonth'].transform('min')
```

```python
#cohort index : the month difference between invoice month and cohort month for each row.
def get_date(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day
invoice_year, invoice_month, _ = get_date(df, 'InvoiceMonth')
cohort_year, cohort_month, _ = get_date(df, 'CohortMonth')
year_diff = invoice_year - cohort_year
month_diff = invoice_month - cohort_month
df['CohortIndex'] = year_diff * 12 + month_diff + 1
```

```python
df.head(10)
```

```python
#No:of active users
cohort_data = df.groupby(['CohortMonth', 'CohortIndex'])['CustomerID'].apply(pd.Series.nunique).reset_index()
cohort_count = cohort_data.pivot_table(index = 'CohortMonth',
                                       columns = 'CohortIndex',
                                       values = 'CustomerID')
cohort_count
```

```python
#The 13th column and 2010–12–1 row means 260 users out of 948 users were still buying things after the 13 months from 2010/12/1.
```

```python
cohort_size = cohort_count.iloc[:,0]
print(cohort_size)
```

```python
retention = cohort_count.divide(cohort_size, axis = 0)
retention.round(3) * 100
```

# HeatMap to represent the retention

```python

plt.figure(figsize = (10,8))
plt.title('Cohort Analysis - Retention Rate')
sns.heatmap(data = retention, 
            annot = True, 
            fmt = '.0%', 
            vmin = 0.0,
            vmax = 0.5,
            cmap = "YlGnBu")
plt.show()
```

# #2nd Cohort: Average Quantity Sold

```python

cohort_data2 = df.groupby(['CohortMonth', 'CohortIndex'])['Quantity'].mean().reset_index()
average_quantity  = cohort_data2.pivot_table(index = 'CohortMonth',
                                            columns = 'CohortIndex',
                                       values = 'Quantity').round(1)
average_quantity
```

```python
plt.figure(figsize = (11,9))
plt.title('Cohort Analysis - Average Quantity')
sns.heatmap(data = average_quantity, 
            annot = True, 
            cmap = "BuGn")
plt.show()
```

# 3rd Cohort: Average Sales

```python
df['TotalSale'] = df['Quantity'] * df['UnitPrice']
cohort_data3 = df.groupby(['CohortMonth', 'CohortIndex'])['TotalSale'].mean().reset_index()
average_sales  = cohort_data3.pivot_table(index = 'CohortMonth',
                                           columns = 'CohortIndex',
                                      values = 'TotalSale').round(1)
average_sales
```

```python
plt.figure(figsize = (11,9))
plt.title('Cohort Analysis - Average Sales')
sns.heatmap(data = average_sales, 
            annot = True, 
            cmap = "Blues")
plt.show()
```
